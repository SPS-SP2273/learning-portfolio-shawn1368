## Problem 2.4

```python
def factorial(n):
    if n == 0:
        return 1
    
    return n * factorial(n - 1)

factorial(5)
```

## Problem 2.5

```python
# For loop + appending in front
def reverse_list(list):
    reversed_list = []
    for i, value in enumerate(list):
        reversed_list = [(i, value)] + reversed_list
    
    return reversed_list

reverse_list([1, 2, 3, 4, 5])
```

## Problem 2.6

```python
def mat_mul(A, B):
    A_row, A_col = A.shape
    B_row, B_col = B.shape
    
    result = np.empty((A_row, B_col))
    
    for i in range(A_row):
        for j in range(B_col):
            result[i, j] = np.sum(A[i, :] * B[:, j])
            
    return result

C = np.arange(1, 5).reshape(2, 2)
D = np.array([[5, 2], [6, 4]])
print(np.matmul(C, D))
print(mat_mul(C, D))
```

## Problem 2.7

```python
def my_argmax(a):
    max_ind = 0
    max_value = a[0]
    for i in range(len(a)):
        if a[i] > max_value:
            max_value = a[i]
            max_ind = i
            
    return max_ind

def my_where(condition, x, y):
    return np.array([x[i] if condition[i] else y[i] for i in range(len(x))])

a = np.arange(10)
print(np.where(a < 5, a, 10*a))
print(my_where(a < 5, a, 10 * a))

print(np.argmax(np.arange(5)))
print(my_argmax(np.arange(5)))
```
